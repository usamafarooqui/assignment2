```python
import pandas as pd
```

```python
poke = pd.read_csv('Pokemon.csv')
```

```python
poke.head()
```

```python
speed_mean = poke['Speed'].mean()
```

```python
speed_mean
```

```python
def set_speed(val):
    if val < speed_mean:
        return "Low speed"
    elif val == speed_mean:
        return "equal speed"
    else:
        return "high speed"
```

```python
poke['High_Low_Speed'] = poke['Speed'].apply(set_speed)
```

```python
poke
```

```python
hp_mean = poke['HP'].mean()
```

```python
hp_mean
```

```python
def set_hp(val):
    if val < hp_mean:
        return "Low HP"
    elif val == hp_mean:
        return "equal Hp"
    else:
        return "High Hp"
```

```python
poke['High_low_hp'] = poke['HP'].apply(set_hp)
```

```python
poke
```
